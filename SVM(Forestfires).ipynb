Data Visualization of the given data set:-
    
Analysis of the Burned Area:-

```python
plt.figure(figsize=(20,8))
print("Skew: {}".format(forest_data['area'].skew()))
print("Kurtosis: {}".format(forest_data['area'].kurtosis()))
ax = sns.kdeplot(forest_data['area'],shade=True,color='deeppink')
plt.xlabel('Area in hectare',color='green',fontsize=16)
plt.ylabel('density of forest fire',color='red',fontsize=16)
plt.title('Forest Fire  Density  Vs  Area Burnt',color='blue',fontsize=20)
plt.xticks([i for i in range(0,1200,50)])
plt.show()
```

The area burned is highly skewed with a value of 12.84 and huge kurtosi value of 194.14.

Forest fire do not cover large area and damage area under 50 hac.

Separating Categorical Columns from the dataset:-

```python
Scc = forest_data.drop(columns='area')
cat_columns = Scc.select_dtypes(include='object').columns.tolist()
```

Separating Numerical Columns from the dataset:-

```python
num_columns = Scc.select_dtypes(exclude='object').columns.tolist() 
```

Analizing the Categorical Columns:-

```python
plt.figure(figsize=(16,10))
for i,col in enumerate(cat_columns,1):
    plt.subplot(2,2,i)            
    sns.countplot(data=Scc,y=col)  
    plt.subplot(2,2,i+2)
    forest_data[col].value_counts().plot.bar() 
    plt.ylabel(col)
    plt.xlabel('% distribution per category')
plt.show()
```

High Number of the forest fires happens in the month of August and September.

Analysis of Forest fire Damage of different Months and Days.

```python
def area_cat(area):            # damage category based on amount of area burned.
    if area == 0.0:
        return "No damage"
    elif area <= 1:
        return "low"
    elif area <= 25:
        return "medium"
    elif area <= 100:
        return "high"
    else:
        return "very high"

forest_data['damage_category'] = forest_data['area'].apply(area_cat)

for col in cat_columns:      
    cross = pd.crosstab(index=forest_data['damage_category'],columns=forest_data[col],normalize='index')
    cross.plot.barh(stacked=True,rot=40,cmap='plasma')
    plt.xlabel('% distribution per category', color = 'green')
    plt.xticks(np.arange(0,1.1,0.1))
    plt.title("Forestfire damage each {}".format(col))
plt.show()
```

Very High Damages occurs in july, August, september. And on saturday reported the most.
