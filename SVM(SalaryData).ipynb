Dropping the Column 'Age':-

```python
train_data=train_data.drop(['age'],axis=1)
train_data.head()
```

```python
test_data=test_data.drop(['age'],axis=1)
test_data.head()
```

Creating a function for Categorical data to Numerical data for training the data:-

```python
def object_to_int(train_data):
    if train_data.dtype == "object":
        train_data = LabelEncoder().fit_transform(train_data)
    return train_data
```

```python
train_data = train_data.apply(lambda x: object_to_int(x))
train_data.head()
```

```python
train_data.describe(include='all')
```

```python
x=train_data.iloc[:,:-1]
y=train_data.iloc[:,-1]
x
```

```python
y
```

```python
x_train, x_test,y_train,y_test = train_test_split(x,y, test_size=0.2,random_state=40)
```

```python
print(x_test.shape,x_train.shape,y_train.shape,y_test.shape)
```

```python
sc_x = StandardScaler()
x_train = sc_x.fit_transform(x_train)
x_test = sc_x.transform(x_test)
```

Now, Applying Support Vector Machine for Classification:-

```python
classifire = SVC(kernel='linear',random_state=0)
classifire.fit(x_train,y_train)
```

Predicting the test results:-

```python
y_pred = classifire.predict(x_train)
```

```python
y_pred
```

```python
classifier = SVC(kernel = 'rbf')
classifier.fit(x_train, y_train)

y_pred=classifier.predict(x_test)

print('Accuracy score  with default rbf kernal')
print(metrics.accuracy_score(y_test,y_pred))
```

```python
classifier = SVC(kernel = 'rbf', gamma=14, C=3,random_state=0)
classifier.fit(x_train, y_train)

y_pred=classifier.predict(x_test)

print('Accuracy score on test data with default rbf linear kernal')
print(metrics.accuracy_score(y_test,y_pred))
```

```python
print('Accuracy score with linear kernal')
print(metrics.accuracy_score(y_test,y_pred))
```

The final accuracy is 85% with the rbf.

Creating a function for Categorical data to Numerical data for testing the data:-

```python
def object_to_int(test_data):
    if test_data.dtype == "object":
        test_data = LabelEncoder().fit_transform(test_data)
    return test_data
```

```python
test_data=test_data.apply(lambda x: object_to_int(x))
test_data.head()
```

```python
test_data.describe(include='all')
```

```python
x=test_data.iloc[:,:-1]
y=test_data.iloc[:,-1]
x
```

```python
y
```

```python
x_train, x_test,y_train,y_test = train_test_split(x,y, test_size=0.2,random_state=40)
```

```python
print(x_test.shape,x_train.shape,y_train.shape,y_test.shape)
```

```python
sc_x = StandardScaler()
x_train = sc_x.fit_transform(x_train)
x_test = sc_x.transform(x_test)
```

Again, Applying Support Vector Machine for Classification:-

```python
classifire = SVC(kernel='linear',random_state=0)
classifire.fit(x_train,y_train)
```

Predicting the test results:-

```python
y_pred = classifire.predict(x_test)
```

```python
y_pred
```

```python
print('Accuracy score with linear kernal')

print(metrics.accuracy_score(y_test,y_pred))
```

```python
classifier = SVC(kernel = 'rbf')
classifier.fit(x_train, y_train)

y_pred=classifier.predict(x_test)

print('Accuracy score  with default rbf kernal')
print(metrics.accuracy_score(y_test,y_pred))
```

```python
classifier = SVC(kernel = 'rbf', gamma=12, C=4,random_state=0)
classifier.fit(x_train, y_train)

y_pred=classifier.predict(x_test)

print('Accuracy score on test data with default rbf linear kernal')
print(metrics.accuracy_score(y_test,y_pred))
```

```python
svc=SVC(kernel='poly',degree=4)
svc.fit(x_train,y_train)

y_pred=svc.predict(x_test)
print('Accuracy score with poly kernal')
print(metrics.accuracy_score(y_test,y_pred))
```

Final accuracy is 83% with rbf.
